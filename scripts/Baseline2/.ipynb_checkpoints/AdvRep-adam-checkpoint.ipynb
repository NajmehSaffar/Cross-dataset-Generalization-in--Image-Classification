```python
# Imports
from __future__ import print_function, division

from PIL import Image
import numpy as np
import time
import os
import sys
import cv2
import pickle
import torch
import torch.optim as optim
import torch.nn as nn
import torch.nn.functional as F
import torch.optim.lr_scheduler as lr_scheduler
import torchvision.utils as utils
from torch.utils.data import Dataset, DataLoader
from torch.utils.data.sampler import SubsetRandomSampler
from torchvision import transforms, datasets, models
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn import metrics
import matplotlib.pyplot as plt
import pandas as pd

from util import *
from model_AdvRep import *

def set_option():
    # Arguments for Training
    opt = {
        "device": torch.device("cuda" if torch.cuda.is_available() else "cpu"),
        
        #"root_dir": '/home/nsaffar/Documents/Mitacs/Dataset',
        #"root_dir": 'C:/Users/saffarn/Desktop/Mitacs/Dataset',
        "root_dir": '/home/ashraflab/Documents/Mitacs/Dataset',
    
        "model_type": 'vgg',
        "model_depth": 16,
        "batch_size": 256,   
        "epochs": 60,
        "num_workers": 4 * torch.cuda.device_count(), 

        # optimization
        "learning_rate": 1e-2,
        #"lr_decay_epochs": '700,800,900',  # where to decay lr, can be a list
        "lr_decay_rate": 0.1,               # decay rate for learning rate
        "weight_decay": 1e-1,
        "momentum": 0.9,
        "dampening": 0.0,
        #"nesterov": True,
        
        "im_size": 32,#128,                      # parameter for RandomResizedCrop
        "n_classes": 1,
        "n_channels": 1,
        "n_datasets": 3
    }
    
    opt["transform"] = transforms.Compose([
        transforms.ToTensor(),
        transforms.ToPILImage(),
        transforms.Resize((opt["im_size"],opt["im_size"])),
        transforms.Grayscale(num_output_channels=opt["n_channels"]),
        transforms.ToTensor(),
        transforms.Normalize((0.5), (0.5))
    ])
    
    return opt


class CustomImageDataset(Dataset):
    def __init__(self, dataframe, img_dir, transform = None):
        
        self.dataframe = dataframe
        self.img_dir = img_dir
        self.transform = transform
        
        self.list_names = dataframe['list_name'].values.tolist()
        self.img_names = dataframe['file_name'].values.tolist()
        self.Y = dataframe['Beluga Present'].values.tolist()
        self.A = dataframe['dataset_membership'].values.tolist()

    def __len__(self):
        return len(self.Y)

    def __getitem__(self, idx):
        
        y = self.Y[idx]
        a = self.A[idx]
        
        img = Image.open(os.path.join(self.img_dir, self.list_names[idx], self.img_names[idx])) 
        if self.transform is not None:
            img = self.transform(img)
            
        imgname = self.img_names[idx]
        
        return img, imgname, y, a
    

class ConcatDataset(Dataset):
    def __init__(self, *datasets):
        self.datasets = datasets

    def __getitem__(self, i):
        return tuple(d[i] for d in self.datasets)

    def __len__(self):
        return min(len(d) for d in self.datasets)    
    
    
def load_dataset(opt):
    
    print('\nloading the dataset ...\n')

    # ------------------------------------- Study 1 -------------------------------------
    df_S1 = pd.read_csv(os.path.join(opt["root_dir"], "Good Quality Photos List.csv")) 
    df_S1['dataset_membership'] = 0
    df_S1['list_name'] = "Good_Quality_Photos"

    number_of_images = len(df_S1['file_name'])
    indices_a, indices_b = train_dev_test_split_indices(number_of_images, train_split=0.5, test_split=0.5)

    df_S1a = df_S1.loc[df_S1.index.isin(indices_a)]
    df_S1b = df_S1.loc[df_S1.index.isin(indices_b)]
    
    print(" ----------------------------------------------------------------------")
    print(" --------------------------- Intersection S1 --------------------------")
    print(" ----------------------------------------------------------------------")
    intersect_dfa_dfb = pd.merge(df_S1a, df_S1b, how='inner', on=['file_name'])
    print(intersect_dfa_dfb)
    
    
    train_indices, valid_indices = train_dev_test_split_indices(number_of_images, train_split=0.9, test_split=0.1)
    df_S1_train = df_S1.loc[df_S1.index.isin(train_indices)]
    df_S1_valid = df_S1.loc[df_S1.index.isin(valid_indices)]

    # ------------------------------------- Study 2 -------------------------------------
    df_S2 = pd.read_csv(os.path.join(opt["root_dir"], "Bad Quality Photos List.csv")) 
    df_S2['dataset_membership'] = 1 
    df_S2['list_name'] = "Bad_Quality_Photos"
    
    number_of_images = len(df_S2['file_name'])
    indices_a, indices_b = train_dev_test_split_indices(number_of_images, train_split=0.5, test_split=0.5)

    df_S2a = df_S2.loc[df_S2.index.isin(indices_a)]
    df_S2b = df_S2.loc[df_S2.index.isin(indices_b)]
    
    print(" -----------------------------------------------------------------------")
    print(" --------------------------- Intersection S2 ---------------------------")
    print(" -----------------------------------------------------------------------")
    intersect_dfa_dfb = pd.merge(df_S2a, df_S2b, how='inner', on=['file_name'])
    print(intersect_dfa_dfb)
    
    
    train_indices, valid_indices = train_dev_test_split_indices(number_of_images, train_split=0.9, test_split=0.1)

    df_S2_train = df_S2.loc[df_S2.index.isin(train_indices)]
    df_S2_valid = df_S2.loc[df_S2.index.isin(valid_indices)]
    
    # ------------------------------------- Study 3 -------------------------------------
    df_S3 = pd.read_csv(os.path.join(opt["root_dir"], "Half Half Photos List.csv")) 
    df_S3['dataset_membership'] = 2
    df_S3['list_name'] = "Half_Half_Photos"

    number_of_images = len(df_S3['file_name'])
    indices_a, indices_b = train_dev_test_split_indices(number_of_images, train_split=0.5, test_split=0.5)

    df_S3a = df_S3.loc[df_S3.index.isin(indices_a)]
    df_S3b = df_S3.loc[df_S3.index.isin(indices_b)]
    
    print(" -----------------------------------------------------------------------")
    print(" --------------------------- Intersection S3 ---------------------------")
    print(" -----------------------------------------------------------------------")
    intersect_dfa_dfb = pd.merge(df_S3a, df_S3b, how='inner', on=['file_name'])
    print(intersect_dfa_dfb)
    
    
    train_indices, valid_indices = train_dev_test_split_indices(number_of_images, train_split=0.9, test_split=0.1)

    df_S3_train = df_S3.loc[df_S3.index.isin(train_indices)]
    df_S3_valid = df_S3.loc[df_S3.index.isin(valid_indices)]
    
    print(" --------------------------- Size of Datasets --------------------------- ")
    print("S1 nfiles:", len(df_S1), ", S2 nfiles:", len(df_S2), ", S3 nfiles:", len(df_S3))
    print('\ndone ...\n')

    
    return [df_S1_train, df_S1_valid, df_S1b, df_S1a], [df_S2_train, df_S2_valid, df_S2b, df_S2a], [df_S3_train, df_S3_valid, df_S3b, df_S3a]


def train(models, train_loader, criterion, optimizer, epoch, device, parameters):
    
    criterion_voi, criterion_dm, criterion_AE = criterion
    model_voi, model_dm, model_enc, model_dec_h = models

    losses_voi = AverageMeter()
    losses_dm = AverageMeter()
    losses_h = AverageMeter()
    losses = AverageMeter()

    model_voi.train()
    model_dm.train()
    model_enc.train()
    model_dec_h.train()
    for batch, data in enumerate(train_loader):
            
        images, imagenames, labels, dataset_memberships = data
        
        Y_df, A_df = pd.DataFrame(labels, columns =['Y']), pd.DataFrame(dataset_memberships, columns =['A'])

        #A0_Y0_wt = 1 / (np.sum(Y_df.iloc[A_df.index[A_df['A'] == 0].tolist()] == 0)[0] + 1)
        #A0_Y1_wt = 1 / (np.sum(Y_df.iloc[A_df.index[A_df['A'] == 0].tolist()] == 1)[0] + 1)
        #A1_Y0_wt = 1 / (np.sum(Y_df.iloc[A_df.index[A_df['A'] == 1].tolist()] == 0)[0] + 1)
        #A1_Y1_wt = 1 / (np.sum(Y_df.iloc[A_df.index[A_df['A'] == 1].tolist()] == 1)[0] + 1)
        #sample_weight = ( A0_Y0_wt * np.multiply(1 - dataset_memberships, 1 - labels) 
        #                + A0_Y1_wt * np.multiply(1 - dataset_memberships, labels) 
        #                + A1_Y0_wt * np.multiply(dataset_memberships, 1 - labels) 
        #                + A1_Y1_wt * np.multiply(dataset_memberships, labels))
        #wts = sample_weight.to(device)
        
        images, target, A = images.to(device), labels.to(device), dataset_memberships.to(device)
        
        target = target.view(target.shape[0],1).float()
        
        # =================== Backward variable_of_interest ====================
        setGradFlags(model_voi, model_dm, model_enc, model_dec_h, True, False, True, True)
            
        Z = model_enc(images)
        X_h = model_dec_h(Z, A)
        out_voi= model_voi(Z)
        out_dm = model_dm(Z)
        
        # Calculate the batch's loss
        loss_h = criterion_AE(X_h, images)
        loss_voi = criterion_voi(out_voi, target)
        loss_dm = criterion_dm(out_dm, A.long())
        #loss_dm = torch.sum((loss_dm * wts))
        
        loss = (loss_h 
                + parameters.get('beta', 1.0) * loss_voi 
                - parameters.get('gamma', 1.0) * loss_dm)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
            
        # Update meters 
        losses_voi.update(loss_voi.item(), 1)
        losses_h.update(loss_h.item(), 1)
        losses.update(loss.item(), 1)
        
        # =================== Backward dataset_membership ====================
        for i in range(5):
            
            setGradFlags(model_voi, model_dm, model_enc, model_dec_h, False, True, False, False)

            Z = model_enc(images)
            X_h = model_dec_h(Z, A)
            out_voi= model_voi(Z)
            out_dm = model_dm(Z)

            # Calculate the batch's loss
            loss_h = criterion_AE(X_h, images)
            loss_voi = criterion_voi(out_voi, target)
            loss_dm = criterion_dm(out_dm, A.long())
            #loss_dm = torch.sum((loss_dm * wts))

            loss = -(loss_h 
                    + parameters.get('beta', 1.0) * loss_voi 
                    - parameters.get('gamma', 1.0) * loss_dm)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            # Update meters 
            losses_dm.update(loss_dm.item(), 1)
    
        
    print(f'Training Process is running: {epoch}/{opt["epochs"]} | Total Loss: {losses.avg} \
    | variable_of_interest loss: {losses_voi.avg} | dataset_membership loss: {losses_dm.avg} \
    | dec_h loss: {losses_h.avg} ')
    
    return losses.avg, losses_voi.avg, losses_dm.avg, losses_h.avg



def evaluate(models, loader, device, str_name):
    correct = 0
    total = 0
    
    model_voi, model_dm, model_enc, model_dec_h = models
    
    # Initialize the prediction and label lists(tensors)
    lbllist=torch.zeros(0,dtype=torch.long, device='cpu')
    outlist=torch.zeros(0,dtype=torch.long, device='cpu')
    predlist=torch.zeros(0,dtype=torch.long, device='cpu')

    model_voi.eval()
    model_dm.eval()
    model_enc.eval()
    model_dec_h.eval()
    with torch.no_grad():
        for i, data in enumerate(loader):
            
            images, imagenames, labels, dataset_memberships = data
            images, target = images.to(device), labels.to(device)
            
            target = target.view(target.shape[0],1).float()
            
            Z = model_enc(images)
            output = model_voi(Z)
            
            predict = torch.round(torch.sigmoid(output.data))

            total += output.size(0)
            correct += torch.eq(predict, target).sum().double().item()

            # Append batch prediction results
            lbllist = torch.cat([lbllist, target.view(-1).cpu()])
            outlist = torch.cat([outlist, output.view(-1).cpu()])
            predlist = torch.cat([predlist, predict.view(-1).cpu()])

    #print('\nAccuracy of the network on test images: %d %%\n' % (100 * correct / total))
    plot_CONFMAT(lbllist.numpy(), predlist.numpy(), str_name)
    roc_auc = plot_ROC(lbllist.numpy(), outlist.numpy(), str_name)
    plot_histogram(lbllist, outlist, str_name)
    
    return roc_auc


def set_loader(df_train, df_valid, df_test_b, df_test_a, opt, proportions):
    
    train_data = CustomImageDataset(df_train, opt["root_dir"], opt['transform'])
    valid_data = CustomImageDataset(df_valid, opt["root_dir"], opt['transform'])
    test_data_b = CustomImageDataset(df_test_b, opt["root_dir"], opt['transform'])
    test_data_a = CustomImageDataset(df_test_a, opt["root_dir"], opt['transform'])
    
    sampler = ProportionalBatchSampler(train_data, opt["batch_size"], proportions)

    train_loader = torch.utils.data.DataLoader(train_data, batch_sampler=sampler, num_workers=opt["num_workers"], pin_memory=True)
                                               #batch_size=opt["batch_size"], num_workers=opt["num_workers"], 
                                               #shuffle=True, pin_memory=True)
     
    valid_loader = torch.utils.data.DataLoader(valid_data, batch_size=opt["batch_size"], num_workers=opt["num_workers"], 
                                              shuffle=True, pin_memory=True)
    
    test_loader_b = torch.utils.data.DataLoader(test_data_b, batch_size=opt["batch_size"], num_workers=opt["num_workers"], 
                                              shuffle=True, pin_memory=True)
    
    test_loader_a = torch.utils.data.DataLoader(test_data_a, batch_size=opt["batch_size"], num_workers=opt["num_workers"], 
                                              shuffle=True, pin_memory=True)
    
    return train_loader, valid_loader, test_loader_b, test_loader_a 


def set_model(opt):#, class_weights):
    
    # ------- Autoencoder -------------------
    model_enc = Encoder(im_size=opt["im_size"], n_channels=opt["n_channels"])
    model_dec_h = Decoder_h(im_size=opt["im_size"], n_channels=opt["n_channels"], n_classes=2)

    # ------- Variable of Interest Classification --------
    model_voi = NaiveNetwork(n_classes=opt["n_classes"])

    # ------- Dataset Membership Classification --------
    model_dm = NaiveNetwork(n_classes=opt["n_datasets"]-1) # one is left for the testing
    
    #if torch.cuda.device_count() > 1:
    #    print("Let's use", torch.cuda.device_count(), "GPUs!")
        # dim = 0 [30, xxx] -> [10, ...], [10, ...], [10, ...] on 3 GPUs
    #    model_enc = nn.DataParallel(model_enc)
    #    model_dec_h = nn.DataParallel(model_dec_h)
    #    model_voi = nn.DataParallel(model_voi)
    #    model_dm = nn.DataParallel(model_dm)

    model_enc.to(opt["device"])
    model_dec_h.to(opt["device"])
    model_voi.to(opt["device"])
    model_dm.to(opt["device"])
   
    criterion_voi = nn.BCEWithLogitsLoss()#pos_weight=class_weights)
    criterion_dm = nn.CrossEntropyLoss()#reduction='none')
    criterion_AE = nn.MSELoss()
    
    # ------- Optimizer ----------------------
    optimizer = optim.Adam([
        {'params': model_voi.parameters()},
        {'params': model_dm.parameters()},
        {'params': model_enc.parameters(), 'lr': 1e-4},
        {'params': model_dec_h.parameters(), 'lr': 1e-4}
        ], lr=opt["learning_rate"])
    
    models = [model_voi, model_dm, model_enc, model_dec_h]
    criterions = [criterion_voi, criterion_dm, criterion_AE]
    
    return models, criterions, optimizer


def train_model(opt, model, train_loader, test_loader, criterion, optimizer, str_name):

    parameters = {"beta":1.0, "gamma":1.0, "m":1, "n":1}
    
    losses, losses_voi, losses_dm, losses_h = [], [], [], []
    for epoch in range(opt["epochs"]):
        
        loss, loss_voi, loss_dm, loss_h = train(model, train_loader, criterion, optimizer, 
                                                epoch, opt["device"], parameters)
        losses.append(loss)
        losses_voi.append(loss_voi)
        losses_dm.append(loss_dm)
        losses_h.append(loss_h)
        
    plot_loss([losses, losses_voi, losses_dm, losses_h], 
              ['combined loss', 'variable_of_interest_C loss', 'dataset_membership_C loss', 'dec_h loss'], 
              ['green', 'blue', 'red', 'orange'], str_name)
    
    plot_loss([losses], ['combined loss'], ['green'], str_name)
    plot_loss([losses_voi], ['variable_of_interest_C loss'], ['blue'], str_name)
    plot_loss([losses_dm], ['dataset_membership_C loss'], ['red'], str_name)
    plot_loss([losses_h], ['dec_h loss'], ['orange'], str_name)
    
    return losses


if __name__ == '__main__':
    
    opt = set_option()
    print("Device: ", opt["device"])

    [df_S1_trn, df_S1_val, df_S1b, df_S1a], [df_S2_trn, df_S2_val, df_S2b, df_S2a], [df_S3_trn, df_S3_val, df_S3b, df_S3a]  = load_dataset(opt)

    print("df_S1a=0 ", sum(df_S1a['Beluga Present']==0), " --- df_S1a=1 ", sum(df_S1a['Beluga Present']==1), " --- df_S1b=0 ", sum(df_S1b['Beluga Present']==0), " --- df_S1b=1 ", sum(df_S1b['Beluga Present']==1))
    print("df_S2a=0 ", sum(df_S2a['Beluga Present']==0), " --- df_S2a=1 ", sum(df_S2a['Beluga Present']==1), " --- df_S2b=0 ", sum(df_S2b['Beluga Present']==0), " --- df_S2b=1 ", sum(df_S2b['Beluga Present']==1))
    print("df_S3a=0 ", sum(df_S3a['Beluga Present']==0), " --- df_S3a=1 ", sum(df_S3a['Beluga Present']==1), " --- df_S3b=0 ", sum(df_S3b['Beluga Present']==0), " --- df_S3b=1 ", sum(df_S3b['Beluga Present']==1))
```
